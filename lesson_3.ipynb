# 2. Напишите код на Python, реализующий построение графиков:
- окружности,
- эллипса,
- гиперболы.

```python
x1 = []
x2 = []
y1 = []
y2 = []

for i in range(1001):
    r = 1000
    x1.append(-i)
    x2.append(i)
    y1.append(np.sqrt(r ** 2 - i ** 2))
    y2.append(-np.sqrt(r ** 2 - i ** 2))
    
plt.plot(x1, y1, color='b')
plt.plot(x1, y2, color='b')
plt.plot(x2, y2, color='b')
plt.plot(x2, y1, color='b')
plt.axis('scaled')
plt.show()
```

```python
# Формула: x=ρcos(φ) y=ρsin(φ)

import math

def polar(r, p):
    x = r*math.cos(p)
    y = r*math.sin(p)
    return x,y

polar(3, 60)
```

# 4. Задание (в программе)
- exp(x) + x∙(1 – y) = 1  
- y = x2 – 1

```python
x = np.linspace(-2, 3, 201)
plt.plot(x, (np.exp(x) + x - 1)/x)
plt.plot(x, x**2 - 1)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return (x**2 - 1 - y, np.exp(x) + x*(1 - y) - 1)

x1, y1 =  fsolve(equations, (2, 4))

              
print (x1, y1)
```
